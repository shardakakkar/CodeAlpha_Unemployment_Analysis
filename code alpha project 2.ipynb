```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df['Date'].dtype
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
df.rename(columns={
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate'
}, inplace=True)

```

```python
df.info()
```

```python
df[['Unemployment_Rate', 'Estimated Employed', 'Labour_Participation_Rate']].describe()
```

```python
plt.figure()
plt.plot(df['Date'], df['Unemployment_Rate'])
plt.title("Unemployment Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure()
sns.lineplot(x='Date', y='Unemployment_Rate', hue='Area', data=df)
plt.title("Unemployment Rate: Rural vs Urban")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
df['Period'] = df['Year'].apply(lambda x: 'Pre-Covid' if x < 2020 else 'Covid & Post-Covid')
```

```python
plt.figure()
sns.boxplot(x='Period', y='Unemployment_Rate', data=df)
plt.title("Impact of Covid-19 on Unemployment Rate")
plt.show()
```

```python
covid_df = df[df['Year'] >= 2020]

plt.figure()
plt.plot(covid_df['Date'], covid_df['Unemployment_Rate'])
plt.title("Unemployment Rate During Covid-19 Period")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Month', y='Unemployment_Rate', data=df)
plt.title("Monthly Seasonal Unemployment Pattern")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
yearly_avg = df.groupby('Year')['Unemployment_Rate'].mean()

plt.figure()
yearly_avg.plot()
plt.title("Average Yearly Unemployment Rate")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure()
sns.barplot(x='Region', y='Unemployment_Rate', data=df)
plt.xticks(rotation=90)
plt.title("Region-wise Average Unemployment Rate")
plt.show()
```
